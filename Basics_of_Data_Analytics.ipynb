4.What is Prompt Engineering, and explain any two types of prompts used in Generative AI?


*   Prompt Engineering is the process of designing and refining the input (called a prompt) given to a generative AI model to get accurate, relevant, and high-quality outputs.
It involves choosing the right words, structure, and context so the AI understands exactly what the user wants.



*   Two Types of prompts:

Instruction-Based Prompt:-This type gives the AI a clear command or task to perform.

Example: “Summarize this paragraph in two sentences.”

Use: Commonly used for tasks like summarization, translation, or writing assistance.

Few-Shot Prompt:-This prompt provides the AI with a few examples before asking it to perform a similar task.

Example 1: Input – ‘Hot’, Output – ‘Cold’  
Example 2: Input – ‘Day’, Output – ‘Night’  
Now, Input – ‘Up’, Output – ?

Use: Helps the model understand the pattern or context before generating an answer.






5.What are the key differences between a Business Analyst and a Data Analyst in terms of roles,
responsibilities, and focus?


* A Business Analyst (BA) focuses on understanding business needs and identifying solutions to improve processes and achieve organizational goals. Their primary role is to act as a bridge between business stakeholders and technical teams. They gather and document business requirements, analyze workflows, and recommend process improvements or system changes. Business Analysts mainly deal with understanding what the business needs and ensuring that the solutions developed meet those requirements. They often use tools such as MS Excel, Jira, Trello, Microsoft Visio, and PowerPoint to manage and present their work.



*  On the other hand, a Data Analyst (DA) focuses on collecting, cleaning,
analyzing, and interpreting data to help organizations make data-driven decisions. Their main responsibility is to uncover patterns, trends, and insights from large datasets. Data Analysts use tools like SQL, Excel, Python or R, Power BI, and Tableau to perform analysis and create dashboards or reports. Their goal is to help businesses make evidence-based decisions and improve performance through data insights.

8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.


*   Descriptive analytics - "What happened?".It is the early stage of data analytics where organization use past data to understand the historical trends.




*   Diagnostic analytics - "Why did it happen?".In this stage, the root cause of each pattern is identified.


*   Predictive analytics - "What will happen?".In this stage,the forecast is predicated by using historical data and statistical models.


*   Prescriptive analytics - "How can we make it happen?"In this stage,the recommend actions is taken that can lead to desired results.


*   Generative AI Analytics – “What can we create or automate?”.In this stage,we can use AI like GPT.Generative AI can not only analyze data but also create reports, summaries, and visualizations automatically.
