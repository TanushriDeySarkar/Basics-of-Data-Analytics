**1.Explain the difference between Relative , Absolute and Mixed Cell Referencing.**


*   **Relative cell Referencing**:A relative reference changes automatically when you copy or drag the formula to another cell.

  Example:Formula for B2 -> B2+10, for B3,it becomes ->b3+10


*   **Absolute cell Referencing**:An absolute reference never changes, even when you copy the formula.
It is locked using $ symbols.

Example:Formula for B2-> B2* $D$1,for B3,B3*$D$1 (refers to the same D1 always)


*   **Mixed cell Referencing**:Mixed reference locks either the row OR the column, but not both.
Example: $A2 → $A2 → $A2 (column fixed)
